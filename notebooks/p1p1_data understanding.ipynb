```python
###import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###display setup
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-whitegrid')

###load dataset
df = pd.read_csv('Housing Market Analysis/train.csv')

###Preview first few rows
df.head()
```

```python
###shape of the dataset
print("Dataset Shape:", df.shape)

###column names
print("\nColumns:\n", df.columns.tolist())

###basic info about data types and missing values
df.info()
```

```python
df.describe().T #descriptive statistics for numerical columns
```

```python
### percentage of missing values per column
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_percentage = (missing/len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing, 'Percentage (%)': missing_percentage})
missing_df.head(10)
```

```python
### examine salesprice distr
plt.figure(figsize=(8,5))
sns.histplot(df['SalePrice'], bins=40, kde=True)
plt.title("Distribution of Sales Price")
plt.xlabel("Sales Price")
plt.ylabel("Frequecy")
plt.show()
```

```python
### correlation overview
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr[['SalePrice']].sort_values(by='SalePrice', ascending = False), annot = True, cmap = 'YlGnBu')
plt.title("Feature Correlation w Sale Price")
plt.show()
```

```python
### Phase 1 Summary - Data Understanding
-Dataset contains 1460 rows and 81 columns
-Target variable: SalePrice (continuous numerical)
- Strongly right-skewed price distribution
- Features with high correlation: OverallQual, GrLiveArea, GarageCars, TotalBsmtSF
-Several columns have missing values (>20%), e.g., PoolQC, MiscFeature
- Next phase: Handle missing values and prepare for analysis
```
